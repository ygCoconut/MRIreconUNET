```python
import glob
import scipy.io as spio
import torch
import numpy as np
```

```python

DATA_DIR = ('C:/Users/littl/Documents/PythonScripts\
/reconproject_data/train_data')

INPUT_NAME = 'NUFFT_img_acq_cropped_norm_cropped'
TARGET_NAME = 'img_Cartesian_acq_cropped_norm_cropped'
```

```python


def dataList(data_dir = DATA_DIR):
    # retrieve list containing all the .mat files of the data_dir folder
    print(data_dir + "/*.mat")
    datadir_list = glob.glob(data_dir + "/*.mat")

    return datadir_list


data_list = dataList()

# print(data_list)
print(len(data_list))

def dataLoad(data_list, input_name = INPUT_NAME, target_name = TARGET_NAME):

    # How to use spio.loadmat:
    # http://www.blogforbrains.com/blog/2014/9/6/loading-matlab-mat-data-in-python
    print(len(data_list))
    for i in range(len(data_list)):
        data_dictionary = spio.loadmat(data_list[i])
        patient_rad = data_dictionary[input_name]
        patient_cart = data_dictionary[target_name]
        

        # HxWxCxN --> NxCxHxW with N the number of slices
        patient_rad = np.transpose(patient_rad, (3, 2, 0, 1))
        patient_cart = np.transpose(patient_cart, (3, 2, 0, 1))
        
        # float64 --> float32
        patient_rad = np.float32(patient_rad)
        patient_cart = np.float32(patient_cart)
        
        filename = data_list
        filename[i] = filename[i].replace('.mat','')
        filename[i] = filename[i].replace(str(DATA_DIR+ '\\' ) ,'')
        print("length patient_rad", len(patient_rad))
#       Give names for each slice of each patient and save them separately
        for j in range(len(patient_rad)):
#             np.savez(str(data_list[i] + "_RADIAL_" + "slice" + j, patient_rad[i])
#             np.savez(str(data_list[i] + "_CARTESIAN_" + "slice" + j, patient_cart[i])
            print("PATIENT:", i, "\t\tSLICE:", j)
            print(str(data_list[i] + "_RADIAL_" + "slice" + str(j)))
            print(patient_rad[j].shape)
#             print(patient_rad[j].dtype)


        print(np.shape(patient_rad))
        print("\n")

dataLoad(data_list)
```
